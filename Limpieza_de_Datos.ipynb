```python
from hashlib import md5


def hashit(val):
    if isinstance(val, float): 
        return str(val)
    return md5(val.encode('utf-8')).hexdigest()
```

```python
#despersonalizar los datos

dataset['user.name'] = dataset['user.name'].apply(hashit)
dataset['user.screen_name'] = dataset['user.screen_name'].apply(hashit)
```
